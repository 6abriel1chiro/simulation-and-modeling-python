```python
from MatModels import *
import numpy
import matplotlib.pyplot as plt
```

```python
# modelo SIR
# N = S(t)+I(t)+R(t)
def deriv(t, y, N, beta, gamma):
    dSdt = -beta * y[0] * y[1] / N
    dIdt = beta * y[0] * y[1] / N - gamma * y[1]
    dRdt = gamma * y[1]
    return numpy.array([dSdt, dIdt, dRdt])
```

```python
# using scipy
# Total population, N.
N = 1000
# Initial number of infected and recovered individuals, I0 and R0.
I0, R0 = 1, 0
# Everyone else, S0, is susceptible to infection initially.
S0 = N - I0 - R0
# Contact rate, beta, and mean recovery rate, gamma, (in 1/days).
beta, gamma = 0.2, 1./10 
# Initial conditions vector
y0 = numpy.array([S0, I0, R0])
# Integrate the SIR equations over the time grid, t.


Y,t = get_euler(0, 160, y0,deriv,0.1,N, beta, gamma)
```

```python
plt.figure(facecolor='w')
plt.plot(t,Y[:,0]/1000, label = "SUSCEPTIBLES")
plt.plot(t,Y[:,1]/1000, label = "INFECTADOS")
plt.plot(t,Y[:,2]/1000, label = "RECUPERADOS")
plt.legend()
plt.grid()
```
