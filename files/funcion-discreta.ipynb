```python
def function(x):
  if(x>=0.5 and x<1.0):
    return 2
  else:
    return 1
```

```python
yPoints=[]
x = np.linspace(0.0, 2.0, 41)

for i in x:
    y = function(i)
    yPoints.append(y)
  

plt.plot(x, yPoints)
plt.show()
```
