```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# de la clase pasada
nx = 41
l = 2.0
dx =l/(nx-1)
nt = 25
dt= 0.01
```

```python
x = np.linspace(0, l, nx)
```

```python
u0 = np.ones(nx)
u0
```

```python
index=np.where(np.logical_and(x>= 0.5, x<=1.0))
index
```

```python
u0[index] = 2
u0
```

```python
plt.plot(x,u0)
plt.grid()
plt.xlim(0.0,l)
plt.ylim(0.0,2.5)
```

```python
#ui ^n+1 = ni ^n + c*dt/dx(ui^n-ui-1^n)
uf = u0.copy()
c = 1

for n in range (1, nt):
  un = uf.copy()
  for i in range (1,nx):
    uf[i] = un[i]-c* dt / dx * (un[i] - un[i-1])
```

```python
plt.plot(x,u0, label = "inicial")
plt.plot(x,uf, label = "final")
plt.grid()
plt.xlim(0.0,l)
plt.ylim(0.0,2.5)
```

```python
def calcularUf(nx,l,c,dt,nt):
  x = np.linspace(0, l, nx)
  dx =l/(nx-1)
  u0 = np.ones(nx)
  index=np.where(np.logical_and(x>= 0.5, x<=1.0))
  u0[index] = 2
  uf = u0.copy()
  for n in range (1, nt):
    un = uf.copy()
    for i in range (1,nx):
      uf[i] = un[i]-c* dt / dx * (un[i] - un[i-1])

  plt.plot(x,u0, label = "inicial")
  plt.plot(x,uf, label = "final")
  plt.grid()
  plt.xlim(0.0,l)
  plt.ylim(0.0,2.5)
```

```python

nx = 101
l = 2.0
nt = 25
dt= 0.002
calcularUf(nx,l,c,dt,nt)
```
