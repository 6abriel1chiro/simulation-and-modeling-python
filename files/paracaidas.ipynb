```python
import matplotlib.pyplot as plt
import numpy
```

```python
def get_logistico(t, P, k,N):
  dPdt = k*P*(1 - P/N)
  return dPdt
```

```python
#presente = 0 
# direccion = f(y,t)
def get_euler(t0, tf, presente, direccion, dt, *args):  #t0, tf, presente, direccion, dt , *args
  values = []
  tiempos= []
  while t0 < tf:
    presente = presente + dt * direccion(t0,presente, *args)
    values.append(presente)
    tiempos.append(t0)
    t0 = t0 + dt
  return values,tiempos
```

```python
def get_paracaidas(t,v,k,m,g):
  dvdt = g-(k/m)*v
  return dvdt
```

```python
res = get_euler(0, 1000,200,0.01, get_logistico,0.03,5000)
plt.plot(res[1], res[0], c='red', label="Simulacion")
plt.grid()
plt.xlabel("tiempo")
plt.ylabel("poblacion")
plt.legend()
```

```python
#
res = get_euler(0, 30,0,0.01, get_paracaidas,12.5,68,9.81)
plt.plot(res[1], res[0], c='red', label="Simulacion")
plt.grid()
plt.xlabel("tiempo")
plt.ylabel("velociad")
plt.legend()
```

```python
def ecosistema(t, ci, a, b, c, d):
  dcdt = a * ci[0] - b * ci[0] * ci[1] # ci es un vector
  dzdt = -c * ci[1] + d * ci[0] * ci[1]
  return numpy.array([dcdt, dzdt])
```
