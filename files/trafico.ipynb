```python
def rhoSemaforo(x,rhoMax):
  if x>0 and x<2:
    return rhoMax* x/2
  elif x>2 and x<4:
    return 0
  else:
    return 0

```

```python


def getRhos(nx,rhoMax):
  yPoints=[]
  x = np.linspace(0.0, l, nx)
  for i in x:
    y = rhoSemaforo(i,rhoMax)
    yPoints.append(y)
  return np.array(yPoints)



```

```python
#f = v * p
#v = vmax(1- p/pmax)
# F = p*v

def flujo(rho,rhoMax, vMax):
  v = vmax*(1-rho/rhoMax)
  F= rho*v
  return F
```

```python
#discretizacion del modelo (foward, t, ) 
# dp / dt + dF/dx = 0
# pi^n+1 - pi^n /nampat

#nx = 81 #nro intervalos
#l = 4.0
#dx = l /(nx-1)
#nt = 30
#vmax = 1.0
#rhoMax = 10.0
#x = np.linspace(0.0, l, nx)

def solucion(nx,nt, dt,myRhos,*args):
  x = np.linspace(0, l, nx)
  u0 = np.ones(nx)
  u0= myRhos
  uf = u0.copy()
  for n in range (1, nt):
    un = uf.copy()
    for i in range (1,nx):
      uf[i] = un[i]-1* dt / dx * (un[i] - un[i-1])

  plt.plot(x,u0, label = "inicial")
  plt.plot(x,uf, label = "final")
  plt.grid()
  plt.xlim(0.0,l)
  plt.ylim(0.0,2.5)
  
```
