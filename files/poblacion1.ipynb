```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
data = np.loadtxt('/content/drive/MyDrive/Modelado/clase05_poblacion/poblacion.csv',delimiter=',')
t_real = data[:, 0]-1790
p_real = data[:, 1]

pl.scatter(t_real, p_real)
pl.grid()
pl.xlabel("tiempo")
pl.ylabel("poblacion real")
```

```python
def getPobliacion(t,p0,k):
  p = p0 * np.exp(k*t)
  return p
```

```python
k = 1/10 * np.log(5.3/3.9)
k
```

```python
getPobliacion(30,3.9,k)
```

```python
t = np.linspace(0,200,100)
t
```

```python
p=getPobliacion(t,3.9,k)
p
```

```python
# grafica de la simulación del modelo
pl.plot(t,p, c='red', label = 'simulacion')
pl.scatter(t_real, p_real, c='blue', label="Poblacion real")
pl.grid()
pl.xlabel("time")
pl.ylabel("population")
pl.legend()
```
